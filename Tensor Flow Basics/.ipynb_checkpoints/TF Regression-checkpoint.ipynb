```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import tensorflow as tf
```

```python
x_data = np.linspace(0.0,10.0,1000000)
```

```python
noise = np.random.randn(len(x_data))
```

```python
print(noise)
```

```python
print(x_data)
```

```python
# y = mx+b
# b =5
```

```python
y_true = (0.5 * x_data) + 5 +noise
```

```python
x_df = pd.DataFrame(data = x_data,columns=['X Data'])
```

```python
y_df = pd.DataFrame(data=y_true,columns=['Y'])
```

```python
y_df.head()
```

```python
my_data = pd.concat([x_df,y_df],axis=1)
```

```python
my_data.head()
```

```python
my_data.sample(n=250).plot(kind='scatter',x='X Data',y='Y')
```

```python
batch_size = 8
```

```python
np.random.randn(2)
```

```python
m = tf.Variable(0.17)
b =tf.Variable(1.12)
```

```python
xph = tf.placeholder(tf.float32,[batch_size])
```

```python
yph = tf.placeholder(tf.float32,[batch_size])
```

```python
y_model = m*xph + b
```

```python
error = tf.reduce_sum(tf.square(yph-y_model))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.001)
train = optimizer.minimize(error)
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    
    sess.run(init)
    
    batches = 100000
    
    for i in range(batches):
        
        rand_ind = np.random.randint(len(x_data),size=batch_size)
        
        feed = {xph:x_data[rand_ind],yph:y_true[rand_ind]}
        
        sess.run(train,feed_dict=feed)
        
    model_m,model_b = sess.run([m,b])
```

```python
model_m
```

```python
model_b
```

```python
y_hat = x_data*model_m + model_b
```

```python
my_data.sample(250).plot(kind = 'scatter',x='X Data',y="Y")
plt.plot(x_data,y_hat,'r')
```

```python
## TF ESTIMATOR
```

```python
feat_cols = [tf.feature_column.numeric_column('x',shape=[1])]
```

```python
estimator = tf.estimator.LinearRegressor(feature_columns=feat_cols)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_eval,y_train,y_eval = train_test_split(x_data,y_true,test_size =0.3,random_state = 101)
```

```python
print(x_eval.shape)
```

```python
input_func = tf.estimator.inputs.numpy_input_fn({'x':x_train},y_train,batch_size=8,num_epochs=None,shuffle=True)
```

```python
train_input_func = tf.estimator.inputs.numpy_input_fn({'x':x_train},y_train,batch_size=8,num_epochs=1000,shuffle=False)
```

```python
eval_input_func = tf.estimator.inputs.numpy_input_fn({'x':x_eval},y_eval,batch_size=8,num_epochs=1000,shuffle=False)
```

```python
estimator.train(input_fn=input_func,steps=1000)
```

```python
train_metrics = estimator.evaluate(input_fn=train_input_func,steps=1000)
```

```python
eval_metrics = estimator.evaluate(input_fn=eval_input_func,steps=1000)
```

```python
print('TRAINING DATA METRICS')
print(train_metrics)
```

```python
print('EVAL DATA METRICS')
print(eval_metrics)
```

```python
data_predict = np.linspace(0,10,10)
input_fn_predict = tf.estimator.inputs.numpy_input_fn({'x':data_predict},shuffle=False)
```

```python
list(estimator.predict(input_fn=input_fn_predict))
```

```python
predicitons = []

for pred in estimator.predict(input_fn=input_fn_predict):
    
    predicitons.append(pred['predictions'])
```

```python
predicitons
```

```python
my_data.sample(250).plot(kind = 'scatter',x='X Data',y="Y")
plt.plot(data_predict,predicitons,'r*')
```
